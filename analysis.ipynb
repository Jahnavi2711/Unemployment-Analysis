```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Data Cleaning

```python
df = pd.read_csv("Unemployment in India.csv")
df1 = pd.read_csv("Unemployment_Rate_upto_11_2020.csv")
```

```python
print(df.columns)
print(df1.columns)
```

```python
print(df1.head())
```

```python
df.columns = df.columns.str.strip()
df1.columns = df1.columns.str.strip()

df["Estimated Employed Rate (%)"] = df["Estimated Labour Participation Rate (%)"] -df["Estimated Unemployment Rate (%)"]
df1["Estimated Employed Rate (%)"] = df1["Estimated Labour Participation Rate (%)"] -df1["Estimated Unemployment Rate (%)"]

print(df.head())
print(df1.head())
```

```python
df["Date"] = pd.to_datetime(df["Date"])
df1["Date"] = pd.to_datetime(df1["Date"])


df['Year']= df["Date"].dt.year
df1['Year']= df1["Date"].dt.year

df["Month_num"] = df["Date"].dt.month
df["Month"] = df["Date"].dt.strftime('%B')

df1["Month_num"] = df1["Date"].dt.month
df1["Month"] = df1["Date"].dt.strftime('%B')

print(df.head())
print(df1.head())
```

```python
print(df["Year"].unique())
print(df.isnull().sum())
print(df1.isnull().sum())
```

```python
df.dropna(inplace =True)
print(df.isnull().sum())

numeric_col = df.select_dtypes(include = [np.number])
infinite = np.isinf(numeric_col).sum()
print(infinite)
```

```python
df.drop(columns=["Frequency","Estimated Employed"],inplace = True)
df1.drop(columns=["Frequency","Estimated Employed"],inplace = True)
```

```python
print(df.head())
print(df1.head())

df.to_csv("new Unemployment in India.csv",index = False)
df1.to_csv("new Unemployment_Rate_upto_11_2020",index = False)
```

```python
df['Period'] = 'Before Covid'
df1['Period'] = 'During Covid'

# Combine the two DataFrames
combined_df = pd.concat([df, df1])
```
